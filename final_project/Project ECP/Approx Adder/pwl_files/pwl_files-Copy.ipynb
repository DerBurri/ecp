```python
import csv

def change_csv_value(filename, row, column, new_value):
    """
    Modifies a specific value in a CSV file.

    Args:
        filename (str): The path to the CSV file.
        row (int): The row index (zero-based).
        column (int): The column index (zero-based).
        new_value: The new value to replace the existing one.
    """

    data = []
    with open(filename, 'r', newline='') as csvfile:
        reader = csv.reader(csvfile)
        for i, row_data in enumerate(reader):
            if i == row:
                row_data[column] = new_value
            data.append(row_data)

    with open(filename, 'w', newline='') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerows(data)


#a0 900m - -1
#a1 900m - -1
#a2 900m - -1
#b0 900m - -1
#b1 900m - -1
#b2 900m - -1
#c 900m - -1
#w0 -1 - 900m
#w1 -1 - 900m

new_value_a0 = "900m"
new_value_a1 = "900m"
new_value_a2 = "900m"
new_value_b0 = "900m"
new_value_b1 = "900m"
new_value_b2 = "900m"
new_value_c = "900m"
new_value_w0 = "-1"
new_value_w1 = "-1"

# Example usage17:
filename = 'a0.csv'
row = 5  # Remember rows are zero-indexed (5th row is index 4)
column = 1  # Columns are also zero-indexed 


change_csv_value(filename, row, column, new_value_a0)

# Example usage16:
filename = 'a0.csv'
row = 6  # Remember rows are zero-indexed (5th row is index 4)
column = 1  # Columns are also zero-indexed 

change_csv_value(filename, row, column, new_value_a0)

# Example usage:
filename = 'a1.csv'
row = 49  # Remember rows are zero-indexed (5th row is index 4)
column = 1  # Columns are also zero-indexed 


change_csv_value(filename, row, column, new_value_a1)

# Example usage15:
filename = 'a1.csv'
row = 50  # Remember rows are zero-indexed (5th row is index 4)
column = 1  # Columns are also zero-indexed 

change_csv_value(filename, row, column, new_value_a1)


# Example usage14:
filename = 'a2.csv'
row = 93  # Remember rows are zero-indexed (5th row is index 4)
column = 1  # Columns are also zero-indexed 

change_csv_value(filename, row, column, new_value_a2)

# Example usage13:
filename = 'a2.csv'
row = 94  # Remember rows are zero-indexed (5th row is index 4)
column = 1  # Columns are also zero-indexed 

change_csv_value(filename, row, column, new_value_a2)


# Example usage12:
filename = 'b0.csv'
row = 7  # Remember rows are zero-indexed (5th row is index 4)
column = 1  # Columns are also zero-indexed 


change_csv_value(filename, row, column, new_value_b0)

# Example usage11:
filename = 'b0.csv'
row = 8  # Remember rows are zero-indexed (5th row is index 4)
column = 1  # Columns are also zero-indexed 

change_csv_value(filename, row, column, new_value_b0)



# Example usage10:
filename = 'b1.csv'
row = 50  # Remember rows are zero-indexed (5th row is index 4)
column = 1  # Columns are also zero-indexed 


change_csv_value(filename, row, column, new_value_b1)

# Example usage9:
filename = 'b1.csv'
row = 51  # Remember rows are zero-indexed (5th row is index 4)
column = 1  # Columns are also zero-indexed 


change_csv_value(filename, row, column, new_value_b1)

# Example usage8:
filename = 'b2.csv'
row = 95  # Remember rows are zero-indexed (5th row is index 4)
column = 1  # Columns are also zero-indexed 


change_csv_value(filename, row, column, new_value_b2)

# Example usage7:
filename = 'b2.csv'
row = 96  # Remember rows are zero-indexed (5th row is index 4)
column = 1  # Columns are also zero-indexed 


change_csv_value(filename, row, column, new_value_b2)


# Example usage6:
filename = 'c.csv'
row = 21  # Remember rows are zero-indexed (5th row is index 4)
column = 1  # Columns are also zero-indexed 


change_csv_value(filename, row, column, new_value_c)

# Example usage5:
filename = 'c.csv'
row = 22  # Remember rows are zero-indexed (5th row is index 4)
column = 1  # Columns are also zero-indexed 

change_csv_value(filename, row, column, new_value_c)

# Example usage4:
filename = 'w0.csv'
row = 1  # Remember rows are zero-indexed (5th row is index 4)
column = 1  # Columns are also zero-indexed 


change_csv_value(filename, row, column, new_value_w0)

# Example usage3:
filename = 'w0.csv'
row = 2  # Remember rows are zero-indexed (5th row is index 4)
column = 1  # Columns are also zero-indexed 

change_csv_value(filename, row, column, new_value_w0)


# Example usage2:
filename = 'w1.csv'
row = 3  # Remember rows are zero-indexed (5th row is index 4)
column = 1  # Columns are also zero-indexed 


change_csv_value(filename, row, column, new_value_w1)


# Example usage1:
filename = 'w1.csv'
row = 4  # Remember rows are zero-indexed (5th row is index 4)
column = 1  # Columns are also zero-indexed 

change_csv_value(filename, row, column, new_value_w1)
```
